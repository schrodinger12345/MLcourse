```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Position_Salaries.csv")
```

```python
X = df.iloc[: , 1:-1].values
y = df.iloc[: , -1].values
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(n_estimators=2, random_state=0)
regr.fit(X,y)
```

```python
regr.predict([[7]])
```

```python
plt.scatter(X,y)
plt.title("Random Forest Regressor")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.plot(X,regr.predict(X), color = "red")
plt.show()
```

```python
X_new = (np.arange(min(X),max(X),0.1)).reshape(len(X_new),1)
plt.scatter(X,y)
plt.title("Random Forest Regressor")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.plot(X_new,regr.predict(X_new), color = "red")
plt.show()
```
