```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("Position_Salaries.csv")
```

```python
X= df.iloc[: , 1:-1].values
y= df.iloc[: , -1].values
```

```python
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor(random_state=0)
dtree.fit(X,y)
```

```python
dtree.predict([[7.77]])
```

```python
plt.scatter(X,y)
plt.plot(X,dtree.predict(X), color = "red")
plt.show()
```
