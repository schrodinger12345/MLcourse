```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Social_Network_Ads.csv", dtype = np.float32)
```

```python
X = df.iloc[: , :-1].values
y = df.iloc[: , -1].values
X = X.astype(np.float32)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X, y , random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train= sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classr = KNeighborsClassifier(n_neighbors= 5, metric = 'minkowski', p =2)
classr.fit(X_train , y_train)
```

```python
classr.predict(sc.transform([[21,70000]]))
```

```python
y_pred = classr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classr.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
